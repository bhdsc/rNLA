```python
import time
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
```

```python
m, n = 100, 10
X = np.random.normal(size=(m, n))
```

```python
# References
# https://www.stat.berkeley.edu/~mmahoney/f13-stat260-cs294/Lectures/lecture02.pdf
# https://web.stanford.edu/class/ee270/scribes/lecture2.pdf

X_ = X # randomized approximation
error = np.sqrt(np.square(X - X_).sum())
```

```python
# Run experiment over many sizes

metrics = {}
metrics["time"] = []
metrics["error"] = []

dim = 10 ** np.arange(1, 7)
for m in dim:
    t = 0
    e = 0
    metrics["time"].append(t)
    metrics["error"].append(e)
```

```python
plt.plot(dim, metrics["time"])
plt.xlabel("dimension")
plt.ylabel("latency")
plt.show()
```

```python
plt.plot(dim, metrics["error"])
plt.xlabel("dimension")
plt.ylabel("error")
plt.show()
```
